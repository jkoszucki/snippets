```python
from pathlib import Path
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import string
import numpy as np

sns.set(font_scale=2)
work_dir = '/Users/januszkoszucki/MGG Dropbox/Janusz Koszucki/working_dir/DATA/KASPAH_KLEBPAVIAP1'
```

```python
df = pd.read_csv(Path(work_dir, 'confident_orfs.csv'))
df['protein_len'] = np.log10(df['protein_len'])
```

```python

fig, axes = plt.subplots(1,2, sharey=True, figsize=(24,10))
fig.suptitle('Distribution of prophage ORF types and sizes (KASPAH & KLEBPAVIA PT1)')

sns.countplot(data=df, y='status', order=['core', 'core-elongated','subcore', 'subcore-elongated', 'single'],ax=axes[0])
sns.boxplot(data=df, y='status', x='protein_len', order=['core', 'core-elongated','subcore', 'subcore-elongated', 'single'], ax=axes[1])

axes[1].set_xlabel('Log(protein length) [aa]')

axes[0].set_ylabel('')
axes[1].set_ylabel('')
```
